```python
# Main Code, make sure to run the pip installs above first!

import altair as alt

import pandas as pd
from datetime import date as dt

with open(r'/content/Comprehensive IMDb Data.csv') as file:
  df = pd.read_csv(file)
#end with

df = df.drop_duplicates(subset=['title'])

df['emoji'] = [{'Drama': '', 'Action': '', 'Adventure': '',
'Comedy': '', 'Horror': '', 'Biography': '', 'Crime': '',
'Fantasy': '', 'Family': '', 'Animation': '', 'Romance': '',
'Music': '', 'Western': '', 'Thriller': '', 'Sci-Fi': '',
'Mystery': '', 'Sport': '', 'Musical': ''
}[genre] for genre in df['genre']]

df['released'] = pd.to_datetime(df['released'], errors='coerce')
df = df.fillna('1978-01-01')

counts_df = df.groupby('genre').size().reset_index(name='count')

alt.renderers.enable('colab')

movie_chart = alt.Chart(df).mark_text(fontSize = 10).encode(
    x = alt.X('released:T', title= "Release Date (1978 for Null Dates)"),
    y = alt.Y('score:Q', title= "IMDB Score /10"),
    tooltip=['title', 'released:T', 'score:Q', 'rating', 'genre'],
    href = 'url:N',
    text = alt.Text('emoji:N'),
    color=alt.Color('legend_txt:N', legend=alt.Legend(title=None))
).transform_calculate(
    url = 'https://m.imdb.com/title/' + alt.datum.titleId,
    legend_txt = alt.datum.emoji + " - " + alt.datum.genre
).properties(
    width = 800,
    height = 600
).interactive()

movie_chart['usermeta'] = {
        "embedOptions": {
            'loader': {'target': '_blank'}
        }
    }

movie_chart
```
